```python
from nltk.tokenize import word_tokenize
movies_df['original_title'] = movies_df['title'].copy()
movies_df['title'] = movies_df['title'].str.lower()
movies_df['description'] = movies_df['description'].str.lower()
movies_df['genres'] = movies_df['genres'].str.lower()

movies_df['title'] = movies_df['title'].apply(word_tokenize)
movies_df['description'] = movies_df['description'].apply(word_tokenize)
movies_df['genres'] = movies_df['genres'].apply(word_tokenize)
```

```python
import nltk
nltk.download('stopwords')
from string import punctuation
from nltk.corpus import stopwords
stopwords = set(stopwords.words('english') + list(punctuation))
stopwords
```

```python
def remove_stopwords(column):
    return column.apply(lambda col: [word for word in col if word not in stopwords])
movies_df['title'] = remove_stopwords(movies_df['title'])
movies_df['description'] = remove_stopwords(movies_df['description'])
movies_df['genres'] = remove_stopwords(movies_df['genres'])
movies_df.head()
```

```python
movies_df['title'] = movies_df['title'].apply(lambda series : list(set(series)))
movies_df['description'] = movies_df['description'].apply(lambda series : list(set(series)))
movies_df['genres'] = movies_df['genres'].apply(lambda series : list(set(series)))
movies_df.head()
```

```python
import gensim

wv = gensim.models.KeyedVectors.load_word2vec_format("GoogleNews-vectors-negative300.bin", binary=True)
```

```python
vocabulary = wv.index_to_key
vocabulary
```

```python
movies_df['title'] = movies_df['title'].apply(lambda col: [word for word in col if word in vocabulary])
movies_df['description'] = movies_df['description'].apply(lambda col: [word for word in col if word in vocabulary])
movies_df['genres'] = movies_df['genres'].apply(lambda col: [word for word in col if word in vocabulary])
movies_df.head()
```

```python

def preprocess_movie_column(movie, column):
    #print(movie[column])
    movie[column] = movie[column].str.lower()
    movie[column] = movie[column].apply(word_tokenize)
    movie[column] = remove_stopwords(movie[column])
    movie[column] = movie[column].apply(lambda series: list(set(series)))
    movie[column] = movie[column].apply(lambda series: [word for word in series if word in vocabulary])
    return movie

def preprocess(movie):
    movie = preprocess_movie_column(movie, 'title')
    movie = preprocess_movie_column(movie, 'description')
    movie = preprocess_movie_column(movie, 'genres')
    return movie
    
    
def recommend(movies):
    
    #empty_scores_df = pd.DataFrame({'title_score': np.zeros(movies_df.shape[0]), 'description_score': movies_df.shape[0], 'genres_score': movies_df.shape[0]})
    #movies_scores_df = pd.concat([movies_df, empty_scores_df], axis = 'columns')
    movies_scores_df = movies_df.copy()
    titles = set([movie['title'] for movie in movies])
    print(titles)
    cnt = 0
    movies_scores_df['total_score'] = pd.Series(np.zeros(movies_scores_df.shape[0]))
    #print(movies_scores_df.head())
    for movie in movies:
        cnt += 1
        movie_df = pd.DataFrame(movie, index = [0])
        movie_df= preprocess(movie_df)
        #print(np.array(movies_scores_df['title']))
        #print(wv.n_similarity(np.array(movie_df.loc[0]['description']), np.array(movies_scores_df.iloc[0]['description'])))
        movies_scores_df[f'title_score'] = movies_scores_df['title'].apply(lambda title: wv.n_similarity(np.array(movie_df.loc[0, 'title']), np.array(title)) if len(title) > 0 else 0)
        movies_scores_df[f'description_score'] = movies_scores_df['description'].apply(lambda description: wv.n_similarity(np.array(movie_df.loc[0, 'description']), np.array(description)))
        movies_scores_df[f'genres_score'] = movies_scores_df['genres'].apply(lambda genres: wv.n_similarity(np.array(movie_df.loc[0, 'genres']), np.array(genres)))
        #print(movies_scores_df.head())
        movies_scores_df['total_score'] = movies_scores_df['total_score'] + movies_scores_df[f'title_score'] + movies_scores_df[f'description_score'] + movies_scores_df[f'genres_score']
    #print(np.array(movies_scores_df[movies_scores_df.original_title == 'Monster Hunter'].original_title))
    #print(np.array(titles))
    #print(np.isin(np.array(movies_scores_df[movies_scores_df.original_title == 'Monster Hunter'].original_title), np.array(titles)))
    #print(np.isin(np.array(movies_scores_df[movies_scores_df.original_title == 'Monster Hunter'].original_title), titles, assume_unique = True, invert = True))
    movies_scores_df = movies_scores_df.loc[np.invert(movies_scores_df.original_title.isin(titles))]
    return movies_scores_df.sort_values(by = "total_score", ascending = False).head(16)['original_title']
```

```python
from flask import Flask, request, jsonify
from flask_restful import Resource, Api, reqparse
```

```python
parser = reqparse.RequestParser()
class Recommender(Resource):
    def post(self):
        movies = request.get_json()
        return list(recommend(movies)), 200
        
```

```python
app = Flask(__name__)
api = Api(app)
api.add_resource(Recommender, '/recommend')
if __name__ == "__main__":
  app.run(debug=True, use_reloader=False)
```
